```python
def metricesCalculation(y_test, pred):
    tn, fp, fn, tp = confusion_matrix(y_test,pred).ravel()
    acc   = (tp+tn)*100 / (tn+fp+fn+tp)
    sens = (tp * 100)  / (tp+fn)
    fpr   =  (fp*100) / (fp +tn)
    return acc,sens,fpr
```

```python
from scipy.stats import pearsonr


# Only compute pearson prod-moment correlations between feature
# columns and target column
target_col_name = 'Depression_type'
feature_target_corr = {}
dataDF.fillna(method='ffill',inplace=True)

for col in dataDF:
  if(col != 'Timestamp'):
    print(pearsonr(dataDF[col].astype(float), dataDF[target_col_name].astype(float)))
    if target_col_name != col:
      feature_target_corr[col] =  pearsonr(dataDF[col].astype(float), dataDF[target_col_name].astype(float))[0]
print("Feature-Target Correlations")
print(feature_target_corr)
```

```python
thresh = 0.01
selected_feature = []
for key in feature_target_corr:
  print(feature_target_corr[key])
  if(feature_target_corr[key]>thresh):
    selected_feature.append(key)

print(selected_feature)
```

```python
print(len(selected_feature))
for i in selected_feature:
  print(i)
```

```python
def correlation(feature_target_corr, thresh):
  
  selected_feature = []
  for key in feature_target_corr:
    if(feature_target_corr[key]>thresh):
      selected_feature.append(key)
  return selected_feature
```

```python
#selected feature according to pearsn with avg threshold
sums_p = 0
for key in feature_target_corr:
  sums_p = sums_p + feature_target_corr[key]

avg_p = sums_p/len(feature_target_corr)
print(avg_p)
```

```python
print("Selected feature according pearson and avg threshold")
selected_feature_p = correlation(feature_target_corr, avg_p)
print(len(selected_feature_p))
for i in selected_feature_p:
  print(i)
```

```python
from sklearn.metrics import r2_score


# Only compute pearson prod-moment correlations between feature
# columns and target column
target_col_name = 'Depression_type'
feature_target_corr_r2 = {}
dataDF.fillna(method='ffill',inplace=True)

for col in dataDF:
  if(col != 'Timestamp'):
    print(r2_score(dataDF[col].astype(float), dataDF[target_col_name].astype(float)))
    if target_col_name != col:
      feature_target_corr_r2[col] =  r2_score(dataDF[col].astype(float), dataDF[target_col_name].astype(float))
print("Feature-Target Correlations")
print(feature_target_corr_r2)
```

```python
#finding threshold
sums =0
for key in feature_target_corr_r2:
  sums = sums +abs(feature_target_corr_r2[key])
avg = sums/ len(feature_target_corr_r2)
print("Finding threshold according to avg value ")
print(avg)
```

```python
# selected Features according r2
print("Selected Features according to R2")
selected_feature_r2 = correlation(feature_target_corr_r2, -avg)
for i in selected_feature_r2:
  print(i)
```
